# Hata hesaplama

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Bağımsız değişken
X = np.array([5, 15, 25, 35, 45, 55]).reshape((-1, 1))
# Bağımlı değişken
y = np.array([5, 20, 14, 32, 22, 38])

model = LinearRegression()

model.fit(X, y)

print('Katsayılar (Coefficients):', model.coef_)
print('Y-Axis Intercept (b):', model.intercept_)

X_new = np.array([60, 70, 80]).reshape((-1, 1))
y_pred = model.predict(X_new)
print("Tahminler:", y_pred)

plt.scatter(X, y, color='blue', label='Gerçek değerler')
plt.plot(X, model.predict(X), color='red', linewidth=2, label='Fit: y = {:.2f}x + {:.2f}'.format(model.coef_[0], model.intercept_))
plt.scatter(X_new, y_pred, color='green', label='Tahminler')

plt.xlabel('X')
plt.ylabel('y')
plt.title('Basit Lineer Regresyon Modeli')
plt.legend()
plt.show()

y_train_pred = model.predict(X)

# Mean squared error (MSE)
mse = mean_squared_error(y, y_train_pred)

# Coefficient of determination (R^2 Score)
r2 = r2_score(y, y_train_pred)

print("Mean squared error: {:.2f}".format(mse))
print("Coefficient of determination: {:.2f}".format(r2))
```

   * Her x ve y değerine ±3 rastgele değişiklik uygulayarak yeni bir dataset oluşturalım.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Adım 1: 40 rastgele x değeri üret ve y değerlerini hesapla
np.random.seed(0)  # Rastgelelik için bir tohum değeri belirler
x = np.random.randint(0, 100, size=40)  # 0 ile 100 arasında 40 tane rastgele tam sayı
y = 3 * x + 4

# Verileri listeleyelim
dataset = list(zip(x, y))
print("Orijinal Dataset:")
print(dataset)

# Adım 2: Grafik çizelim
plt.scatter(x, y, color='blue', label='Orijinal veriler')
plt.title("Orijinal Dataset Grafiği")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()

# Adım 3: Her x ve y değerine ±3 rastgele değişiklik uygula
x_new = x + np.random.randint(-3, 4, size=x.shape)
y_new = y + np.random.randint(-3, 4, size=y.shape)

# Yeni verileri listeleyelim
new_dataset = list(zip(x_new, y_new))
print("Yeni Dataset (+-3):")
print(new_dataset)

#Yeni dataset üzerinden basit lineer regresyon modeli kurup, modelin formülünü hesaplayalım.
#modeli üç veri üzerinden eğitelim
#yeni formülü hesaplayalım
```
